```python
import csv
import random
import numpy as np
import os
with open('train.tsv', "r", encoding="utf-8") as f:
    reader = csv.reader(f,delimiter = "\t")
    texts = []
    labels = []
    for row in reader:
        text = row[0]
        label = row[1]
        if text == 'text':
            continue
        texts.append(text)
        labels.append(label)

n_cluster = 5
labels_standard = np.unique(labels)
random.shuffle(labels_standard)
labels_train = labels_standard[:n_cluster]
labels_train_path = 'labels_train.npy'
labels_path = 'labels.npy'
np.save(labels_train_path,labels_train)
np.save(labels_path,labels_standard)

labels = np.array(labels)    
res_text = []
res_label = []
res_id = []
supervised_ratio = 1
for label in labels_train:
    num = int(len(labels[labels == label]) * supervised_ratio)
    pos = list(np.where(labels == label)[0])
    res_id.extend(random.sample(pos, num))

random.shuffle(res_id)
for idx in res_id:
    res_text.append(texts[idx])
    res_label.append(labels[idx])

with open('1.tsv',"w", encoding="utf-8") as f:
    writer = csv.writer(f,delimiter = "\t")
    writer.writerow(['text','label'])
    for text,label in zip(res_text, res_label):
        writer.writerow([text,label])
```
